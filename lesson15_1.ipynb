```python
#codespace 
#install -y
#:sudo apt-get update -y
#:sudo apt-get install -y fonts-wqy-zenhei
#:https://askubuntu.com/questions/490829/how-can-i-install-chinese-fonts-on-kubuntu-14-04
#:/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc

import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
#mac: font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)
#mac: font1=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=12)
#window: font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)
font = FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)
figure = plt.figure(figsize=(8,5),dpi=250,facecolor='gray')
Axes = figure.add_subplot(1,1,1)
stu1 = {'':56,'':45, '':89,'':75,'':63}
stu2 = {'':46,'':35, '':75,'':63,'':72}
Axes.plot(stu1.keys(),stu1.values(),'g--o',label='stu1')
Axes.plot(stu2.keys(),stu2.values(),'r-+',label='stu2')
Axes.set_xticklabels(stu1.keys(),fontproperties=font)
Axes.legend()
Axes.set_title("12",fontproperties=font)
Axes.set_xlabel("",fontproperties=font)
Axes.set_ylabel("",fontproperties=font)
for x in Axes.get_xticks():
    value1 = list(stu1.values())[x]
    value2 = list(stu2.values())[x]
    print(x,value1)
    Axes.annotate(str(value1),(x+0.05,value1),fontproperties=font,color='red',
                  arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.95', 
                            color='b'))
    Axes.annotate(str(value2),(x+0.05,value2),fontproperties=font,color='green',
                  arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.95', 
                            color='b'))
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
days = [num for num in range(1,31)]
celsius = [round(random.uniform(24.3,37.5),ndigits=1) for _ in range(1,31)]
figure = plt.figure(figsize=(8,5))
axes = figure.add_subplot()
axes.plot(days,celsius,'r--.',label='September')
axes.legend()
axes.grid()
axes.set_title('weather')
axes.set_xlabel('days')
axes.set_ylabel('celsius')
axes.set_xticks(days)
plt.show()
```
