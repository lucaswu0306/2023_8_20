```python
import requests
#下載動作
ur1 = 'https://opendata.cwb.gov.tw/fileapi/v1/opendataapi/F-C0032-001?Authorization=rdec-key-123-45678-011121314&format=JSON'

response = requests.get(ur1)
print(type(response))

if response.status_code == 200:
    print("下載成功")
else:
    print("下載失敗")
    
```

```python
all_data = response.json()
data_list = all_data['cwaopendata']['dataset']['location']
#print(type(data_list))
weather_list = []
for item in data_list:
    city_item = {}
    city_item['城市'] = item['locationName']
    city_item['起始時間'] = item['weatherElement'][0]['time'][0]['startTime']
    city_item['結束時間'] = item['weatherElement'][0]['time'][0]['endTime']
    city_item['最高溫度'] = item['weatherElement'][1]['time'][0]['parameter']['parameterName']
    city_item['最低溫度'] = item['weatherElement'][2]['time'][0]['parameter']['parameterName']
    city_item['感覺'] = item['weatherElement'][3]['time'][0]['parameter']['parameterName']#dict資料中key 裡面的 key
    weather_list.append(city_item)
weather_list
```

```python
import pandas as pd
df = pd.DataFrame(weather_list)
df['城市']
```
