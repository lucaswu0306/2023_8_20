```python
#標準寫法
import matplotlib.pyplot as plt

# 1.製作figure
fig = plt.figure()

# 2.在1.的figure上製作一個axes
ax_1 = fig.add_subplot(1, 1, 1)

# 3.將圖的數據儲存在axes中
X = [0, 1, 2]
Y = [3, 4, 5]
ax_1.plot(X, Y)

# 4.顯示圖
plt.show()
```

```python
figure = plt.figure()
ax1 = figure.add_subplot(1,3,1)
ax1.set_title('Axes1')
ax2 = figure.add_subplot(1,3,2)
ax2.set_title('Axes2')
ax3 = figure.add_subplot(1,3,3)
ax3.set_title('Axes3')
plt.show()
```

```python
figure = plt.figure(figsize=(8,5),dpi=200)
axes = figure.add_subplot()#預設1,1,1
student = [56, 45, 89, 75,63]
axes.plot(['chinese', 'english', 'math', 'science', 'created'],student)#只寫一個代表y軸，寫兩個時，第一個代表x軸，第二個代表y軸
plt.show()
```

```python
figure = plt.figure(figsize=(8,5),dpi=200)
axes = figure.add_subplot()#預設1,1,1
stu1 = {'chinese':56,'english':45, 'math':89,'science':75,'created':63}
stu2 = {'chinese':46,'english':89, 'math':13,'science':52,'created':22}
axes.plot(stu1.keys(),stu1.values())
axes.plot(stu2.keys(),stu2.values())
plt.show()
```
