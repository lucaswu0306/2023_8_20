```python
import requests
#下載動作
ur1 = 'https://opendata.cwb.gov.tw/fileapi/v1/opendataapi/F-C0032-001?Authorization=rdec-key-123-45678-011121314&format=JSON'

response = requests.get(ur1)
print(type(response))

if response.status_code == 200:
    print("下載成功")
    
```

```python
#存檔

with open("最新的氣象預測.json",mode='w',encoding='utf-8') as file:#json檔案不用寫newline=''
    file.write(response.text)
```
