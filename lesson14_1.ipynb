```python
#將資料結成想要的內容，成為簡單的表格資料
data_list = all_data['cwbopendata']['dataset']['location']
#print(type(data_list))
weather_list = []
for item in data_list:
    city_item = {}
    city_item['城市'] = item['locationName']
    city_item['起始時間'] = item['weatherElement'][0]['time'][0]['startTime']
    city_item['結束時間'] = item['weatherElement'][0]['time'][0]['endTime']
    city_item['最高溫度'] = item['weatherElement'][1]['time'][0]['parameter']['parameterName']
    city_item['最低溫度'] = item['weatherElement'][2]['time'][0]['parameter']['parameterName']
    city_item['感覺'] = item['weatherElement'][3]['time'][0]['parameter']['parameterName']#dict資料中key 裡面的 key
    weather_list.append(city_item)
weather_list
```

```python
import os
import csv
data_list = all_data['cwbopendata']['dataset']['location']
#print(type(data_list))
weather_list = []
for item in data_list:
    city_item = {}
    city_item['城市'] = item['locationName']
    city_item['起始時間'] = item['weatherElement'][0]['time'][0]['startTime']
    city_item['結束時間'] = item['weatherElement'][0]['time'][0]['endTime']
    city_item['最高溫度'] = item['weatherElement'][1]['time'][0]['parameter']['parameterName']
    city_item['最低溫度'] = item['weatherElement'][2]['time'][0]['parameter']['parameterName']
    city_item['感覺'] = item['weatherElement'][3]['time'][0]['parameter']['parameterName']
    weather_list.append(city_item)

root = os.path.abspath('./')
filePath = os.path.join(root,'csv_data','最新的氣象預測.csv')
with open(filePath,mode='w',encoding='utf-8',newline='') as file:
    filenames = ['城市','起始時間','結束時間','最高溫度','最低溫度','感覺']
    csv_dictwriter = csv.DictWriter(file,filenames)
    csv_dictwriter.writeheader()
    csv_dictwriter.writerows(weather_list)

```
