```python
import requests
url = 'https://raw.githubusercontent.com/roberthsu2003/PythonForDataAnalysis/master/pandas%E7%9A%84%E5%9F%BA%E7%A4%8E%E5%8A%9F%E8%83%BD/%E5%90%84%E9%84%89%E9%8E%AE%E5%B8%82%E5%8D%80%E4%BA%BA%E5%8F%A3%E5%AF%86%E5%BA%A6.csv'
response = requests.get(url)
if response.status_code == 200:
    print("下載成功")
else:
    print("下載失敗")
```

```python
r = requests.get(url,stream=True)
with open('各鄉鎮市區人口密度.csv', 'wb') as fd:
    for chunk in r.iter_content(chunk_size=128):
        print(chunk)#print(chunk)输出的是字节数据类型（bytes）。
        fd.write(chunk)
```

```python
#默認情况下是stream=False，他會立即開始下载文件並存放到内存當中，倘若文件過大就會導致内存不足的情况．
#當把get函数的stream参数设置成True时，它不會立即開始下载，當你使用iter_content或iter_lines遍歷内容或訪問内容属性时才開始下载。

url = 'https://doqvf81n9htmm.cloudfront.net/data/crop_article/107759/5db17bf947590.jpg_1140x855.jpg'
r = requests.get(url,stream=True)#因為jpg圖片檔大部份都是二進制，所以在使用open()時要將mode='wb，而這個b 就是binary mode opening-二進制模式
with open('pic.jpg', 'wb') as fd:#w是寫入
    for chunk in r.iter_content(chunk_size=128):#每128bytes為一個chunk
        fd.write(chunk)
```

```python

url = 'https://upload.wikimedia.org/wikipedia/commons/2/2d/Parthenon-Restoration-Nov-2005-a.jpg'
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36'}
r = requests.get(url,stream=True,headers=headers)
with open('pic.jpg', 'wb') as fd:#w是寫入b是bits stream是用bits
    for chunk in r.iter_content(chunk_size=128):
        fd.write(chunk)
```
