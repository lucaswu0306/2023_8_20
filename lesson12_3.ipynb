```python
import os
import csv
root = os.path.abspath('./')
filePath = os.path.join(root,"個股日成交資訊.csv")

with open(filePath,encoding='utf-8',newline='\n') as file:
  csv_dictreader = csv.DictReader(file)
  for item in csv_dictreader:
    print(item['證券名稱'], item['收盤價'])
```
