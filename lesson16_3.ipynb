```python
import numpy as np
scores = np.random.randint(60,100,size=(50,5))
import pandas as pd
students = pd.DataFrame(scores,index=range(1,51),columns=['國文','英文','數學','地理','公民'])
students
#print(type(students))
```

```python
import os
import csv
root = os.path.abspath('./')
filePath = os.path.join(root,'16_3.csv')
with open(filePath,mode='w',encoding='utf-8',newline='') as file:
    csv_writer = csv.writer(file)
    csv_writer.writerows(scores)
```

```python
import os
import csv
root = os.path.abspath('./')
filePath = os.path.join(root,'16_3.csv')
file =  open(filePath,mode='w',encoding='utf-8',newline='')
csv_writer = csv.writer(file)
csv_writer.writerow(['國文','英文','數學','地理','公民'])
csv_writer.writerows(scores)
file.close()
```
