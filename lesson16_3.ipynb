```python
import numpy as np
scores = np.random.randint(60,100,size=(50,5))
import pandas as pd
students = pd.DataFrame(scores,index=range(1,51),columns=['國文','英文','數學','地理','公民'])
students
```

```python
#使用subscript->['字串'],選欄
#使用subscript->[1:5],選列
students['國文']#series
```

```python
students[['國文','數學']]#DataFrame
```

```python
students[0:1]
```

```python
#students.loc[index]
#studnets.iloc[真正的索引值編號]
students.loc[50], students.iloc[49]
```

```python
#loc包含5，iloc不包含5
students.loc[1:5], students.iloc[0:5]
```

```python
#不建議的寫法
one = students.loc[1]
one.sum()
```

```python
students.loc[1].sum()
```

```python
students.sum(axis=1)#axis 中 index=0，column=1
```

```python
#增加DataFrame的欄位
students['總分'] = students.sum(axis=1)
students
```

```python
students[['國文','英文','數學','地理','公民']]
```

```python
#loc是標籤名，跟索引名index的名
students.loc[1:50,['國文','英文','數學','地理','公民']]
```

```python
#iloc是索引值
students['平均']= students.iloc[:,:5].mean(axis=1)#mean是平均
students
```

```python
students.to_csv('學生第一學期分數.csv')
```
