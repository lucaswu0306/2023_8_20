```python
#stack
import numpy as np
import pandas as pd

frame = pd.Series(np.random.randn(9),
          index=[
              ['a', 'a', 'a', 'b', 'b', 'c', 'c', 'd', 'd'],
              [1, 2, 3, 1, 2, 3, 1, 2, 3]
          ])
frame
```

```python
frame['b']
```

```python
frame['b':'c']
```

```python
data = frame.unstack()
data.fillna(0)
```

```python
frame.unstack().stack()
```

```python
frame1 = pd.DataFrame(np.arange(12).reshape((4,3)),
                      index=[['a','a','b','b'],[1, 2, 1, 2]],
                      columns=[['台北','台北','台中'],['Green','Red','Green']])
frame1
```

```python
frame1.index.names = ['key1','key2']
frame1
```

```python
frame1.columns.names = ['縣市','顏色']
frame1
```

```python
frame1['台北']
```

```python
frame1
```

```python
#level可以對調
frame1.swaplevel()
```

```python
frame1.sort_index(level = 0, ascending = False)
```

```python
frame1.sort_index(level=1)
```

```python
frame1.groupby(level=1).sum()
```

```python
frame1.groupby(by = 'key2').sum()
```

```python
frame1.groupby(level=0).sum()
```

```python
frame1.groupby(axis=1,level='顏色').sum()
```
