```python
import numpy as np
import pandas as pd

#美國球賽門票收入
tips = pd.read_csv('tips.csv')
tips.columns = ['總票價', '小費', '吸煙者', '日期', '時間', '大小']
tips['小費比例'] = tips['小費'] / tips['總票價']
tips
```

```python
tips.sort_values(by = '小費比例')[-6:]
```

```python
def top(df, n=5, column = '小費比例'):
    return df.sort_values(by=column)[-n:]
top(tips,n=6)
```

```python
tips.groupby(['吸煙者','日期','時間']).sum()
```

```python
len(tips.groupby('吸煙者'))
```

```python
dict(list(tips.groupby('吸煙者')))
```

```python
tips.groupby('吸煙者').apply(top,n=6)
```
