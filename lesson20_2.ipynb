```python
import pandas as pd
frame = pd.read_excel('學生第一學期分數.xlsx')
#frame1 = frame.rename(columns = {'Unnamed: 0':'學號'})
frame
```

```python
frame1 = frame[['學號','國文','英文','數學','地理','公民']]
frame2 = frame1.set_index('學號')
frame2
```

```python
def description(s:pd.Series) -> pd.Series:
    max = s.max()
    max_value = s[s==max].index[0]
    min = s.min()
    min_value = s[s==min].index[0]
    std = s.std()
    return pd.Series([max, max_value,min,min_value, std],index=['最高分','最高分科目','最低分','最低分分數','標準差'])

frame2.apply(description,axis=1)
```
