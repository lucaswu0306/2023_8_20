```python
import numpy as np
index = np.arange(5)
value = np.array([5,7,3,4,6])*1000
import matplotlib.pyplot as plt
figure = plt.figure(figsize=(8,5))
axes = figure.add_subplot()
axes.bar(index,value)
axes.set_xticks(index,['apple','orange','potato','banana','guava'])
axes.grid(axis='y')
xmin,xmax,ymin,ymax= axes.axis()
print(xmin,xmax,ymin,ymax)
#axes.axis([xmin-1, xmax+1, ymin, ymax+1000])#讓x軸的範圍變大
for i in index:
    print(i,value[i],end=' ')
    axes.annotate(str(value[i]),[i-0.18,value[i]+100])
    
    
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
sdata = pd.Series({'Taipei':35000, 'Taoyuan':71000, 'Taichung':16000, 'Kaohsiung':5000})
figure = plt.figure(figsize=(8,5))
axes = figure.add_subplot()
axes.bar(sdata.index,sdata.values)
axes.grid(axis='y')
xmin,xmax,ymin,ymax=axes.axis()
#print(xmin,xmax,ymin,ymax)
axes.axis([xmin-1,xmax+1,ymin,ymax])
plt.show()
```
