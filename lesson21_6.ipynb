```python
import numpy as np 
import pandas as pd
import requests
url1 = 'https://ods.railway.gov.tw/tra-ods-web/ods/download/dataResource/8ae4cabf6973990e0169947ed32454b9'
response = requests.get(url1)
jsonData = response.json()
jsonData
```

```python
#trnOpDate(乘車日)、staCode(車站代碼)、gateInComingCnt(進站人數)、gateOutGoingCnt(出站人數)
dataFrame = pd.DataFrame(jsonData)
dataFrame.columns = ['乘車日','車站代碼','進站人數','出站人數']
dataFrame
```

```python
dataFrame.to_excel("a.xlsx")
```

```python
url2 = 'https://ods.railway.gov.tw/tra-ods-web/ods/download/dataResource/0518b833e8964d53bfea3f7691aea0ee'
response = requests.get(url2)
stations_json = response.json()
stations_json
```

```python
stations = pd.DataFrame(stations_json,columns=['stationCode','stationName'])
stations.columns = ['車站代碼','車站名稱']
stations
```

```python
mergeframe = pd.merge(dataFrame,stations,on='車站代碼',how='left')#left以左邊為基準所有在做邊的資料都要出現，right以右邊為基準所有在右邊的資料都要出現
mergeframe
```

```python
mergeframe.info()
```

```python
mergeframe['乘車日'] = pd.to_datetime(mergeframe['乘車日'])
```

```python
#進站人數，出站人數是字串
mergeframe.info()
```

```python
def to_float(v):
    print(v)
    return v
mergeframe['進站人數'].map(to_float)
```

```python
def to_float1(v):
    v = int(v)
    if v>1000:
        return "超過1000人"
    else:
        return "少於1000人"
mergeframe['訊息'] = mergeframe['進站人數'].map(to_float1)#dataframe新增欄位: df[''] = series
mergeframe
```

```python
mergeframe.info()
```

```python
mergeframe['進站人數'] = mergeframe['進站人數'].astype(int)
mergeframe['出站人數'] = mergeframe['出站人數'].astype(int)
```

```python
mergeframe.info()
```

```python
def to_int(v):
    v = int(v)
    return v
mergeframe['進站人數'] = mergeframe['進站人數'].map(to_int)
mergeframe
```

```python
mergeframe.info()
```

```python
mergeframe2 = mergeframe.set_index(['乘車日','車站名稱']).drop(columns = ['車站代碼','訊息'])
mergeframe2#index 有3個level 分別為 level = 0, level = 1, level = 2
```

```python
mergeframe3 = mergeframe2.stack().unstack(level = 1)
mergeframe3
mergeframe4 = mergeframe2.stack()
mergeframe4.index.names = ['乘車日','車站名稱','進出']
mergeframe4
```
